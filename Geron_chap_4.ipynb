```python
np.random.seed(42)
m = 100
X = 2*np.random.rand(m,1)
y = 4 + 3*X + np.random.randn(m,1)
```

`np.randn` is a function in the NumPy library that generates random numbers from the standard normal distribution (mean = 0, standard deviation = 1).`np.random.randn(d0, d1, ..., dn)`.  `d0, d1, ..., dn` (optional): Dimensions of the resulting array.

```python
fig = plt.figure()
plt.scatter(X,y)
plt.grid()
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
from sklearn.preprocessing import add_dummy_feature
```

```python
X_b = add_dummy_feature(X) # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y  # the @ operator performs matrix multiplication
```

The function that we used to generate the data is y = 4 + 3x + Gaussian noise.
